```python
c = list(zip(xs, ys))
xs, ys = zip(*c)
```

```python
train_xs = xs[:int(len(xs) * 0.8)]
train_ys = ys[:int(len(xs) * 0.8)]
```

```python
val_xs = xs[-int(len(xs) * 0.2):]
val_ys = ys[-int(len(xs) * 0.2):]   
```

```python
num_train_images = len(train_xs)
num_val_images = len(val_xs)
```

```python
def LoadTrainBatch(batch_size):
    global train_batch_pointer
    x_out = []
    y_out = []
    for i in range(0, batch_size):
        x_out.append(cv2.resize(cv2.imread(train_xs[(train_batch_pointer + i) % num_train_images])[-150:], (200, 66)) / 255.0)
        y_out.append([train_ys[(train_batch_pointer + i) % num_train_images]])
    train_batch_pointer += batch_size
    return x_out, y_out
```

```python
def LoadValBatch(batch_size):
    global val_batch_pointer
    x_out = []
    y_out = []
    for i in range(0, batch_size):
        x_out.append(cv2.resize(cv2.imread(val_xs[(val_batch_pointer + i) % num_val_images])[-150:], (200, 66)) / 255.0)
        y_out.append([val_ys[(val_batch_pointer + i) % num_val_images]])
    val_batch_pointer += batch_size
    return x_out, y_out
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models
def build_model():
    model = models.Sequential()
    # First convolutional layer
    model.add(layers.Conv2D(24, (5, 5), strides=2, activation='relu', input_shape=(66, 200, 3)))
    # Second convolutional layer
    model.add(layers.Conv2D(36, (5, 5), strides=2, activation='relu'))
    # Third convolutional layer
    model.add(layers.Conv2D(48, (5, 5), strides=2, activation='relu'))
    # Fourth convolutional layer
    model.add(layers.Conv2D(64, (3, 3), strides=1, activation='relu'))
    # Fifth convolutional layer
    model.add(layers.Conv2D(64, (3, 3), strides=1, activation='relu'))
    # Flatten the output
    model.add(layers.Flatten())
    # Fully connected layer 1
    model.add(layers.Dense(1164, activation='relu'))
    model.add(layers.Dropout(0.5))
    # Fully connected layer 2
    model.add(layers.Dense(100, activation='relu'))
    model.add(layers.Dropout(0.5))
    # Fully connected layer 3
    model.add(layers.Dense(50, activation='relu'))
    model.add(layers.Dropout(0.5))
    # Fully connected layer 4
    model.add(layers.Dense(10, activation='relu'))
    model.add(layers.Dropout(0.5))
    # Output layer
    model.add(layers.Dense(1))
    model.add(layers.Lambda(lambda x: tf.multiply(tf.math.atan(x), 2)))

    return model
model = build_model()
model.summary()
```

```python
import os
import tensorflow as tf
from tensorflow.keras.callbacks import ModelCheckpoint, TensorBoard

# driving_dataset = "data/driving_dataset"

class Trainer:
    def __init__(self, model, log_dir, l2_norm_const=0.001, learning_rate=1e-4):
        self.model = model
        self.log_dir = log_dir
        self.l2_norm_const = l2_norm_const

        # Compile the model with L2 regularization
        self.model.compile(
            optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate),
            loss=self._custom_loss(),
            metrics=['mae']
        )

        # Prepare directories for saving model and logs
        self._prepare_directories()

        # Set up callbacks
        self.checkpoint_callback = ModelCheckpoint(
            filepath=os.path.join(self.log_dir, "model_{epoch:02d}.keras"),
            save_best_only=True,
            monitor="val_loss",
            verbose=1
        )
        self.tensorboard_callback = TensorBoard(
            log_dir=os.path.join(self.log_dir, "logs"),
            histogram_freq=1
        )

    def _prepare_directories(self):
        """Ensure the log directory exists for saving model and TensorBoard logs."""
        if not os.path.exists(self.log_dir):
            os.makedirs(self.log_dir)
        if not os.path.exists(os.path.join(self.log_dir, 'logs')):
            os.makedirs(os.path.join(self.log_dir, 'logs'))

    def _custom_loss(self):
        """Custom loss function combining MSE and L2 regularization."""
        def loss_fn(y_true, y_pred):
            mse_loss = tf.reduce_mean(tf.square(y_true - y_pred))
            l2_loss = tf.add_n([tf.nn.l2_loss(v) for v in self.model.trainable_variables])
            return mse_loss + self.l2_norm_const * l2_loss
        return loss_fn

    def train(self, epochs, batch_size):
        # Load training and validation data
        train_gen = self._data_generator(LoadTrainBatch, batch_size)
        val_gen = self._data_generator(LoadValBatch, batch_size)

        # Train the model
        self.model.fit(
            train_gen,
            steps_per_epoch=num_train_images // batch_size,
            validation_data=val_gen,
            validation_steps=num_val_images // batch_size,
            epochs=epochs,
            callbacks=[self.checkpoint_callback, self.tensorboard_callback]
        )

    def _data_generator(self, data_loader, batch_size):
        """Generator function for loading batches."""
        while True:
            xs, ys = data_loader(batch_size)
            yield tf.convert_to_tensor(xs, dtype=tf.float32), tf.convert_to_tensor(ys, dtype=tf.float32)

if __name__ == "__main__":
    # Define directories and hyperparameters
    LOGDIR = './save'
    LOGS_PATH = './logs'  # Separate logs path for TensorBoard
    EPOCHS = 30
    BATCH_SIZE = 100

    # Create the trainer and start training
    trainer = Trainer(model, LOGDIR)
    trainer.train(EPOCHS, BATCH_SIZE)

    # Provide instructions to launch TensorBoard
    print("Run the command line:\n" \
          "--> tensorboard --logdir=./model_training/train_steering_angle/logs " \
          "\nThen open http://0.0.0.0:6006/ into your web browser")
```
